```python
import pandas as pd
import seaborn as sns
from ydata_profiling import ProfileReport
```

```python
#Load the dataset 
Tunisair_data = pd.read_csv('Tunisair_flights_dataset.csv')

#Display first 5 rows
Tunisair = Tunisair_data.head() 

Tunisair
```

```python
# Display overview of the data
Tunisair_info = Tunisair_data.info() 

Tunisair_info
```

```python
# Display missing values
Tunisair_missing_values = Tunisair_data.isnull().sum()

Tunisair_missing_values 
```

```python
# Summary statistics for the dataset
Tunisair_statistics_summary = Tunisair_data.describe() 

Tunisair_statistics_summary
```

```python
#Using the first 500 rows for ydata-profiling
Tunisair_500rows = Tunisair_data.head(500) 

#Use ydata-profiling to generate the report
profile = ProfileReport(Tunisair_500rows, title="Tunisair Report")
```

```python
profile.to_notebook_iframe()# Display the report
```
